### 함수의 정의와 기능

#### 함수 정의

    def 함수이름(매개변수1, 매개변수2,...) :
        코드블록
        return
        
    + def - 함수의 시작
    + 함수이름 - 어떤 기능을 하는 함수인지 최대한 나타낼 수 있도록 동사 또는 명사를 사용하여 작명, 변수이름규칙 적용
    + 매개변수(parameter) - 함수에 입력으로 전달하는 값, 전달하는 값을 인자(argument)라고 하기도 함.
    + 콜론, 코드블록(들여쓰기) 
    + return - 기본적으로 함수의 종료를 명시
        return 옆에 값이나 수식이 있다면 해당 값을 호출자(caller)에게 반환
        
    + 함수 호출전에 함수를 정의해야 함 - 정의되지 않은 함수를 호출하면 error발생
    
    * 함수의 이름 :
        - 어떤 기능을 하는 함수인지 이름으로 최대한 나타날 수 있게 해야함
        - 의미와 반대되거나 맞지 않는 이름은 사용금지

#### - def guess_plusORminus(x):

```python

x=int(input('임의의 숫자를 입력하세요: '))
def guess_plusORminus(x):
    
    if x < 0:
        
        return '음수'
    
    else:
       
        return '양수'
        
guess_plusORminus(x)
```

## 사전의 특성
    1) 사전은 dict로 자료형을 정의한다.
    2) 키key와 값value의 쌍으로 항목item이 구성된다. 
    3) 구조 변경이 가능하다 - 키 : 값 항목을 추가, 삭제, 수정할 수 있다. (리스트와 유사, 튜플과 다름)
    4) 순서가 없다 
        -  사전의 키를 정렬할 수는 있으나,  순서가 없다. 즉, 순차자료가 아니다.(non-sequential)
        - 인덱싱(번호매기기) 슬라이싱(일부 자료 추출)을 할 수 없다.
    5) 키는 중복된 값을 갖지 않는다 
        - 값(value)에 해당하는 고유한 키가있다
        - 사전의 키 타입: 정수/실수/문자열/튜플 같은 변경 불가능한 자료형만 허용된다
    6) 값은 중복될 수 있으며 모든 자료형이 가능하다
    * 키의 중복: 허용되나 결과를 장담할 수 없다

## 집합(set)의 개념 및 특성

 + 사전 자료구조에서 값(value)은 버리고 키(key)만 남은 자료구조와 같다. (각 키는 유일해야 함)
 + 수학의 집합 개념을 파이썬에서 이용할 수 있도록 만든 자료구조이다
 + 파이썬은 집합연산이 가능한 집합 set 자료구조를 제공한다. 

 + 수학에서의 집합 속성을 갖는다.( 중복된 값이 없음. 순서 없음. 교집합, 합집합, 차집합 연산 가능)
 + 집합 자료구조의 효용 - 어떤 항목이 존재하는지 여부만 판단할때 유용하다.
  
   
##### 집합의 특성
1. 중복을 허용하지 않는다.
2. 순서가 없다(Unordered)

```python
set1 = {1, 2, 3}

set1
```

```python
set2 = {1, 1, 3, 3, 2}

set2
```

## 집합의 특성  
#### 파이썬 집합의 특성
    - 집합의 항목 순서가 없다 : 인덱싱을 할 수 없다(인덱싱을 하려면 리스트나 튜플로 변환)

    - 집합 항목을 중복할 수 없다.

#### 파이썬 집합의 활용
    - 중복되는 항목을 제거하고, 고유(unique)한 항목으로 재구성한다.
